```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
plt.plot(names, values) 
plt.figure()
```

```python
plt.subplot(2,2,1)
plt.bar(names, values)

plt.subplot(2,2,2)
plt.scatter(names, values)

plt.subplot(2,2,4)
plt.plot(names, values)
```

```python
plt.subplot(3,2,1)
plt.bar(names, values)

plt.subplot(3,2,2)
plt.scatter(names, values)

plt.subplot(3,2,3)
plt.plot(names, values)

plt.subplot(3,2,4)
plt.bar(names, values)
plt.subplot(3,2,5)
plt.scatter(names, values)
plt.subplot(3,2,6)
plt.plot(names, values)

plt.suptitle('Categorical Plotting')
```
